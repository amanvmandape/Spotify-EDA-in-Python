# Spotify Exploratory Data Analysis

### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

### Data Cleaning

```python
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

#### Converting the Release_Date to datetime format

```python
df_tracks['release_date'] = pd.to_datetime(df_tracks.release_date)
```

#### Converting the duration from miliseconds to seconds

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace = True, axis = 1)
```

### Data Analysis

#### 1) Find the least popular 10 songs on spotify

```python
unpopular_df = df_tracks.sort_values('popularity').head(10)
unpopular_df
```

#### 2) Find the descriptive statistics for the data set

```python
df_tracks.describe().transpose()
```

#### 3) Find the top 10 most popular songs on spotify

```python
popular_df = df_tracks.sort_values('popularity', ascending = False).head(10)
popular_df
```

#### 4) Create a corelation map

```python
df_corr = df_tracks.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson', numeric_only = True)
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(df_corr, annot = True, fmt = '.1g', vmin = -1, vmax = 1, center = 0, cmap='inferno', linewidths = 1, linecolor = 'black')
heatmap.set_title('Correlation HeatMap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

#### 5) Create Regression Plots using the Data

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y='loudness', x='energy', color = 'c').set(title = 'Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y='popularity', x='acousticness', color = 'y').set(title = 'Population vs Acousticness Correlation')
```

#### 6) Create distribution Plot for songs released each year

```python
years = df_tracks.release_date.dt.year
sns.displot(years, discrete=True, aspect=2, height=5, kind='hist').set(title="Number of Songs Per Year")
```
